```python
import pandas as pd
import matplotlib as plt 
import seaborn as sns

df = pd.read_csv("premier_league.csv")
print(df.head())
print(df.info())
print(df.describe())

try:
   df["date"] = df["date"].str.replace(r"(st|nd|rd|th)", "", regex=True)
   df["date"] = pd.to_datetime(df["date"], format="%d %B %Y")
   df["clock"] = pd.to_datetime(df["clock"], format="%I:%M%p").dt.time
   df = df.drop(columns=["links"])
except Exception as e:
     print("Error while cleaning:", e)

print(df.info())
print(df.head())
print(df.columns.tolist())
```
